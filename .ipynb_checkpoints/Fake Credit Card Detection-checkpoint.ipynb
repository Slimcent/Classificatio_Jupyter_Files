```python
import urllib.request 
import pandas as pd
import matplotlib.pyplot as plt

my_data = 'https://raw.githubusercontent.com/Slimcent/Machine_Learning_Datasets/main/ML/credit_card_fraud.csv'

def download_read_data(path):
    
    """
    Function to retrieve data from the data paths
    And to read the data as a pandas dataframe
  
    To return the dataframe
    """ 
    # Only retrieve the directry of the data
    
    data_path =  urllib.request.urlretrieve(path)[0]
    data = pd.read_csv(str(data_path))

    return data
```

```python
# Getting train data 

data = download_read_data(my_data)
```

Count: Number of non-null values in the column.

Mean: Average value of the column.

Standard deviation: Measure of how spread out the values are from the mean.

Minimum: Smallest value in the column.

25th percentile (Q1): Value below which 25% of the data points fall.

50th percentile (Median): Value that divides the data into two halves.

75th percentile (Q3): Value below which 75% of the data points fall.

Maximum: Largest value in the column.
