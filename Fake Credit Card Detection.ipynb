```python
import urllib.request 
import pandas as pd

my_data = 'https://raw.githubusercontent.com/Slimcent/Machine_Learning_Datasets/main/ML/credit_card_fraud.csv'

def download_read_data(path):
    
    """
    Function to retrieve data from the data paths
    And to read the data as a pandas dataframe
  
    To return the dataframe
    """ 
    # Only retrieve the directry of the data
    
    data_path =  urllib.request.urlretrieve(path)[0]
    data = pd.read_csv(str(data_path))

    return data
```

```python
# Getting train data 

data = download_read_data(my_data)
```
